# LiteracyPy 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from numpy import nan as NaN


# 2018 Top 20 Most Educated Countries In the World based on https://naijaquest.com/most-educated-countries-in-the-world/
countries = ['Japan', 
             'South Korea', 
             'Republic of Korea',
             'United Kingdom',
             'United Kingdom of Great Britain and Northern Ireland',
             'Singapore',
             'Russia', 
             'Russian Federation',
             'Finland',
             'Canada',
             'Netherlands',
             'Ireland',
             'Israel',
             'China',
             'China, Macao Special Administrative Region',
             'New Zealand',
             'Norway',
             'Belgium',
             'Germany',
             'Denmark',
             'Estonia',
             'United States',
             'United States of America',
             'France',
             'Portugal'] 

# Literacy rate per country, adult total (% of people ages 15 and above)
# source: blob:https://ourworldindata.org/a14b0aa9-8650-44ba-9666-a82b33c3d3f0

literacy_data__to_load = "source_data/cross-country-literacy-rates.csv"
df_literacy_raw = pd.read_csv(literacy_data__to_load)
df_literacy_raw.head()
```

```python
# data munging

# extract the top 20 
df_literacy = df_literacy_raw.loc[df_literacy_raw['Entity'].isin(countries)]

# get values for years 1980 onwards
df_literacy = df_literacy.loc[df_literacy['Year']>1979]


df_literacy.columns = ['Country', 'Code', 'Year', 'Literacy Rate (%)']
df_literacy.drop('Code', axis=1, inplace=True)
df_literacy.head()
```

```python
# pivot data and fill in values for each row
df_literacy_pv = df_literacy.pivot(index='Year',columns='Country',values='Literacy Rate (%)')
df_literacy_pv.fillna(method='ffill', inplace=True)
df_literacy_pv.bfill(axis ='rows', inplace=True) 
df_literacy_pv.head()
```

### Line Graph - Literacy Rate in 1980-2015

```python
# Generate the Plot
x_axis = df_literacy_pv.index
line_width = 2
set_alpha = 1

from itertools import cycle
colors = ['burlywood',
         'dodgerblue',
         'goldenrod',
         'mediumvioletred',
         'black',
         'red',
         'teal',
         'olivedrab',
         'royalblue',
         'deeppink',
         'lightcoral',
         'darkseagreen',
         'rosybrown',
         'cadetblue',
         'c',
         'forestgreen',
         'plum',
         'navy',
         'darkkhaki',
         'yellow']
cycolor = cycle(colors)

fig, ax = plt.subplots(figsize=[10, 12])

# multiple line plot
for country in df_literacy_pv:
    line_color = next(cycolor)
    if country == 'United States':
        line_color = 'darkorange'
        line_width = 6
        set_alpha = 0.5
    ax.plot(x_axis, df_literacy_pv[country], marker='', color=line_color, linewidth=line_width, alpha=set_alpha, label=country)
    # get a random point on the line and add country label
    x = np.random.choice(list(df_literacy_pv.index))
    y = df_literacy_pv.loc[x, country]
    plt.text(x, y, country, color=line_color)
    
plt.tight_layout()
ax.legend(loc='center left', bbox_to_anchor=(0.8, 0.35), ncol=1)
plt.xlabel('Year')
plt.ylabel('Literacy Rate (%)')
plt.title('Literacy Rate (%) over the years 1980-2015', fontsize='16')
plt.grid()
plt.savefig("output_data/literacy_rates.jpg", dpi=200, transparent=True, quality=95, bbox_inches="tight")
plt.show()
```

### Horizontal Bar Function

```python
import matplotlib.ticker as ticker
def plotLiteracyCrossDataHorizontalBar(plotType, df, x, y, fig, year):
    df_sorted = df.sort_values(by=[year])

    if plotType == 1:
        ax = df_sorted[[x, y]].plot(kind='barh', figsize=(8, 14), colormap='Paired', zorder=2, width=0.85)
    else:
        ax = df_sorted[[x, y]].plot(kind='bar', figsize=(8, 14), colormap='Paired', zorder=2, width=0.85)
           
    # Despine
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

    # Switch off ticks
    ax.tick_params(axis="both", which="both", bottom=False, top=False, labelbottom=True, left=False, right=False, labelleft=True)
    for country in ax.get_yticklabels():
        if country.get_text() == 'United States':
            country.set_color("r")
        elif country.get_text() == 'Finland':
            country.set_color("indianred")
        elif country.get_text() == 'Norway':
            country.set_color("salmon")
        elif country.get_text() == 'Estonia':
            country.set_color("lightcoral")
        elif country.get_text() == 'United Kingdom':
            country.set_color("darksalmon")
        elif country.get_text() == 'Russia':
            country.set_color("darkorchid")
            
        
    ax.xaxis.set_label_text("")
    ax.yaxis.set_label_text("")
    
    if plotType == 1:
        for y_loc, x_loc in enumerate(df_sorted[x]):
            plt.annotate(str(round(x_loc, 2)), xy=(x_loc, y_loc), va='center')
        for y_loc, x_loc in enumerate(df_sorted[y]):
            plt.annotate(str(round(x_loc, 2)), xy=(x_loc, y_loc), va='center')
    
    plt.title('Literacy Rate and Education Spend Per Capita) in {}'.format(year), fontsize='16')
    ax.legend(["Literacy Rate", "Education Spend Per Capita"], loc='upper right', bbox_to_anchor=(1.4, 1), ncol=1)
    plt.savefig(fig, dpi=200, transparent=True, quality=95, bbox_inches="tight")
```
