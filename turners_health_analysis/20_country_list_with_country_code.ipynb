```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#import seaborn as sb
```

```python
countries_subset = ['Japan',
            'South Korea', # Korea, Dem. Peopleâ€™s Rep.
            'United Kingdom',
            'Singapore',
            'Russia', # Russian Federation
            'Finland',
            'Canada',
            'Netherlands',
            'Ireland',
            'Israel',
            'China',
            'New Zealand',
            'Norway',
            'Belgium',
            'Germany',
            'Denmark',
            'Estonia',
            'United States',
            'France',
            'Portugal']
```

```python
# File to Load (Remember to change these)
OWID_life_expectancy_csv_path = "data_sources/OWID_life-expectancy.csv"

# Read in the data from the CSV
OWID_life_expectancy_data = pd.read_csv(OWID_life_expectancy_csv_path)

# convert data to a DataFrame
OWID_life_expectancy_df = pd.DataFrame(OWID_life_expectancy_data)

#OWID_life_expectancy_df.head()
#OWID_life_expectancy_df["Entity"].unique()

country_subset_life_expectancy = OWID_life_expectancy_df.loc[(OWID_life_expectancy_df["Entity"].isin(countries_subset)),:]

countries_and_codes_subset = country_subset_life_expectancy[["Entity","Code"]].drop_duplicates().rename(columns={"Entity":"country","Code":"country_code"})

countries_and_codes_subset

# Save path to store output file set in a variable
csv_output = "data_exports/countries_and_codes_20Countries.csv"

# Export file as a CSV, without the Pandas index, 
# but with the header
countries_and_codes_subset.to_csv(csv_output, index=False, header=True)

```
